```python
import pandas as pd

file_path = "Review.csv"
df = pd.read_csv(file_path)

pd.set_option('display.max_colwidth', None)
print(df)
```

```python
df["lowercased"] = df["Review"].str.lower()
df[["Review", "lowercased"]]
```

```python
import re

df["urls_removed"] = df["lowercased"].apply(
    lambda x: re.sub(r'http\S+|www\S+', '', x)
)

df[["lowercased", "urls_removed"]]
```

```python
from bs4 import BeautifulSoup

df["html_removed"] = df["urls_removed"].apply(
    lambda x: BeautifulSoup(x, "html.parser").get_text()
)

df[["urls_removed", "html_removed"]]
```

```python
import re

df["emojis_removed"] = df["html_removed"].apply(
    lambda x: re.sub(r'[^\x00-\x7F]+', '', x)
)

df[["html_removed", "emojis_removed"]]
```

```python
slang_dict = {
    "tbh": "to be honest",
    "omg": "oh my god",
    "idk": "i do not know"
}

def replace_slang(text):
    words = text.split()
    new_words = []
    for word in words:
        if word in slang_dict:
            new_words.append(slang_dict[word])
        else:
            new_words.append(word)
    return " ".join(new_words)

df["slangs_replaced"] = df["emojis_removed"].apply(replace_slang)

df[["emojis_removed", "slangs_replaced"]]
```

```python
contractions = {
    "can't": "cannot",
    "won't": "will not",
    "don't": "do not",
    "didn't": "did not",
    "it's": "it is",
    "i'm": "i am",
    "i've": "i have",
    "i'll": "i will",
    "you're": "you are",
    "they're": "they are",
    "isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not"
}

def replace_contractions(text):
    for c, full in contractions.items():
        text = text.replace(c, full)
    return text

df["contractions_replaced"] = df["slangs_replaced"].apply(replace_contractions)

df[["slangs_replaced", "contractions_replaced"]]
```

```python
import string

df["punctuations_removed"] = df["contractions_replaced"].apply(
    lambda x: x.translate(str.maketrans('', '', string.punctuation))
)

df[["contractions_replaced", "punctuations_removed"]]
```

```python
import re

df["numbers_removed"] = df["punctuations_removed"].apply(
    lambda x: re.sub(r'\d+', '', x)
)

df[["punctuations_removed", "numbers_removed"]]
```

```python
spelling_dict = {
    "battry": "battery",
    "excelent": "excellent",
    "chargin": "charging",
    "wasnt": "was not"
}

def correct_spelling(text):
    words = text.split()
    corrected_words = []
    for word in words:
        if word in spelling_dict:
            corrected_words.append(spelling_dict[word])
        else:
            corrected_words.append(word)
    return " ".join(corrected_words)

df["spelling_corrected"] = df["numbers_removed"].apply(correct_spelling)

df[["numbers_removed", "spelling_corrected"]]
```

```python
basic_stopwords = {
    "the","is","am","are","was","were","a","an","and","or","but","if","to","of","in","on","for","with",
    "this","that","it","i","my","me","we","our","you","your","he","she","they","them","his","her",
    "have","has","had","do","does","did","not","so","too","very","just","ever","be","been"
}

df["stopwords_removed"] = df["spelling_corrected"].apply(
    lambda x: " ".join([w for w in x.split() if w not in basic_stopwords])
)

df[["spelling_corrected", "stopwords_removed"]]
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

df["stemmed_words"] = df["stopwords_removed"].apply(
    lambda x: " ".join([stemmer.stem(word) for word in x.split()])
)

df[["stopwords_removed", "stemmed_words"]]
```

```python
lemmatization_dict = {
    "running": "run",
    "worked": "work",
    "fitting": "fit",
    "comfortable": "comfortable",
    "better": "good"
}

def simple_lemmatize(text):
    words = text.split()
    new_words = []
    for word in words:
        if word in lemmatization_dict:
            new_words.append(lemmatization_dict[word])
        else:
            new_words.append(word)
    return " ".join(new_words)

df["lemmatized"] = df["stopwords_removed"].apply(simple_lemmatize)

df[["stopwords_removed", "lemmatized"]]
```

```python
df.to_csv("Processed_Reviews.csv", index=False)
```
